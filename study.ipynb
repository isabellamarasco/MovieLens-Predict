```python
#import
import pandas as pd
import matplotlib.pyplot as plt

path = '/Users/isabella/Desktop/dataset-ml-25m/'
dataset = pd.read_csv(path + 'dataset.csv', sep=',')

#plot the number of movies for each rating in order to see if the dataset is balanced
dataset['rating'].value_counts().plot(kind='bar', figsize=(15,5), title='Number of movies for each rating')

#save the image in png
plt.savefig('imgs/unbalance.png')
```
